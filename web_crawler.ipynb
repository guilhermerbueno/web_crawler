# Web Crawler ou Web Scraping
Coleta de dados da web, utilizando programas para "raspar" informações destes sites que podem ser utilizadas para futuras análises. Ex: comparar preços, monitorar o tempo, pesquisar oque está em alta no mercado, etc.

## Beautiful Soup
Biblioteca utilizada para formatar um arquivo html

## Requests
Biblioteca utilizada para fazer requisições http


```python
from bs4 import BeautifulSoup
import requests
```

## Parametros
### Nosso crawler terá apenas 3 parâmetros
**URL**: URL da página que desejamos analisar <br/>
**html_object**: Qual o objeto HTML queremos extrair  <br/>
**css_class**: Qual a classe css que o objeto possui  <br/>

```python
#url = "http://www.climatempo.com.br/" <br/>
#html_object = "img" <br/>
#css_class = "_margin-l-10 _margin-r-5" <br/>
```

```python
url = "https://www.infomoney.com.br/cotacoes/petrobras-petr4/"
html_object = "div"
css_class = "value"
```

### Efetua a chamada ao site e salva o retorno na variável html, então o html é formatado utilizando a biblioteca BeautifulSoup e printamos o html

```python
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

### Busca pelo dado desejado no html

```python
obj = soup.find(html_object, class_=css_class)
print(obj.p.string)
```
